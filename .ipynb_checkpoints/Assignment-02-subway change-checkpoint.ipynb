```python
import re
import requests
url = 'http://bj.bendibao.com/ditie/linemap.shtml'
response = requests.get(url)
pattern = re.compile(r'<a href="(\/ditie\/map_\d+\.shtml)" target="_blank">(\w+\(?\w+\)?\w+)</a>')
list1 = pattern.findall(response.text)
```

```python
pattern = re.compile(r'\w+\(?\w+\)?\w+')
string = '今天(天气)不错,今天天气不错'
pattern.findall(string)
```

```python
dict = {}
for i in list1:
    url2, name = i
    url = 'http://bj.bendibao.com/'+url2
    response2 = requests.get(url)
    pattern2 = re.compile(r'<a href="\/ditie\/zd_\w+\.shtml" class="link" target="_blank">(\w+)</a>')
    station_name_list = pattern2.findall(response2.text)
    dict[name] = station_name_list
dict
```
