### 常用语法

1.创建画布和创建子图

plt.figure: 创建一个空白画布，制定画布大小，像素

figure.add_subplot: 创建并选中子图，制定子图行数，列数和选中图片的编号

2.添加画布内容

plt.title: 添加标题，指定标题的名称，位置，颜色，字体大小等参数

plt.xlabel: 添加x轴名称，指定位置，颜色，字体大小等参数

plt.ylabel: 添加y轴名称，指定位置，颜色，字体大小等参数

plt.xlim: 指定图形x轴的范围，只能确定一个数值区间，无法使用字符串标识

plt.ylim: 指定图形y轴的范围，只能确定一个数值区间，无法使用字符串标识

plt.xticks: 指定x轴刻度的数目和取值

plt.yticks: 指定y轴刻度的数目和取值

plt.legend: 指定当前图形图例，指定图例的大小，位置，标签

3.保存与显示图形

plt.savefig: 保存绘制的图形，指定图形的分辨率，边缘颜色等参数

plt.show: 显示图形

```python
%matplotlib inline
```

```python
data = np.arange(0,1.0,0.01)
plt.title('lines')
plt.xlabel('x')
plt.ylabel('y')
plt.xticks([0,0.2,0.4,0.6,0.8,1])
plt.yticks([0,0.2,0.4,0.6,0.8,1])
plt.plot(data,data **2)
plt.plot(data,data **4)
plt.legend(['y=x^2','y=x^4'])
plt.show()
```

```python
rad = np.arange(0,np.pi*2,0.01)
pl = plt.figure(figsize=(8,6),dpi=80)
# 确定画布大小和像素
ax1 = pl.add_subplot(2,1,1)
# 创建一个2行1列的子图，并开始绘制第一幅图
plt.title('lines')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim((0,1))
plt.ylim((0,1))
plt.xticks([0,0.2,0.4,0.6,0.8,1])
plt.yticks([0,0.2,0.4,0.6,0.8,1])
plt.plot(rad,rad **2)
plt.plot(rad,rad **4)
plt.legend(['y=x^2','y=x^4'])
# 第二幅图
pl.add_subplot(2,1,2)
plt.title('sin/cos')
plt.xlabel('rad')
plt.ylabel('value')
plt.xlim((0,np.pi*2))
plt.ylim((-1,1))
plt.xticks([0,np.pi/2,np.pi,np.pi*1.5,np.pi*2])
plt.yticks([-1,-0.5,0,0.5,1])
plt.plot(rad,np.sin(rad))
plt.plot(rad,np.cos(rad))
plt.legend(('sin','cos'))
plt.show()
```

### 设置pyplot的动态rc参数

```python
x = np.linspace(0,4*np.pi)
y = np.sin(x)
plt.plot(x,y,label="$sin(x)$")
plt.title('sin')
plt.show()
```

修改rc参数

```python
plt.rcParams['lines.linestyle'] = '-.'
plt.rcParams['lines.linewidth'] = 3
plt.plot(x,y,label="$sin(x)$")
plt.title('sin')
plt.show()
```

### 常用语法

lines.linewidth: 线条宽度，取0~10之间的数值，默认为1.5

line.linestyle: 线条样式，可取"-" "--" "-." ":"四种，默认为"-"

lines.marker: 线条上点的形状，可取"o" "D" "h" "." "," "S"等20种，默认为None

lines.markersize: 点的大小，取0~10之间的数值，默认为1

### 散点图

表征两特征之间的关系

```python
#  matplotlib.pyplot.scatter(x,y,s=None,c=None,marker=None,cmap=None,norm=None,vmin=None,vmax=None,
#  aipha=None,linewidths=None,verts=None,edgecolors=None,hold=None,data=None,**kwargs)
```

x,y: 接收array.表示x轴和y轴对应的数据，无默认

s: 接受数值或一维的array，制定点的大小，若传入一维array，则表示每个点的大小，默认为None

c: 接收颜色或者一维的array,指定点的颜色，若传入一维array，则表示每个点的颜色，默认为None

marker: 接收特定的string，表示绘制点的类型，默认为None

alpha: 接收0-1的小数，表示点的透明度，默认为None

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = np.loead('../data/国民经济核算季度数据.npz')
name = data['columns']
values = data['values']
plt.figure(figsize=(8,7))
plt.scatter(values[:,0],values[:,2],marker='o')
plt.xlabel('年份')
plt.ylabel('生产总值（亿元）')
plt.ylim((0,225000))
plt.xticks(range(0,70,4),values[range(0,70,4),1],rotation=45)
plt.title('2000-2017年各季度生产总值散点图')
plt.savefig()
plt.show()
```

### 折线图

随时间变化的趋势

matplotlib.pyplot.plot(*args, **kwargs)

x,y: 接收array.表示x轴和y轴对应的数据，无默认

color： 接收特定string，指定颜色，默认None

linestyle: 接收特定string，指定线条类型，默认为'-'

marker: 接收特定string, 表示绘制点的类型，默认None

alpha: 接收0~1的小数，表示点的透明度。默认None

```python
plt.figure(figsize=(8,7))
plt.plot(values[:,0],values[:,2],color='r',linestyle='--')
plt.xlabel('年份')
plt.ylabel('生产总值（亿元）')
plt.ylim((0,225000))
plt.xticks(range(0,70,4),values[range(0,70,4),1],rotation=45)
plt.title('2000-2017年各季度生产总值散点图')
plt.savefig()
plt.show()
```

一次接收多组数据，添加多条折线图，分别定义每条折线的颜色，点的形状和类型

```python
plt.figure(figsize=(8,7))
plt.plot(values[:,0],values[:,3],'bs-',
        values[:,0],values[:,4],'ro-.',
        values[:,0],values[:,5],'gH--')
plt.xlabel('年份')
plt.ylabel('生产总值（亿元）')
plt.ylim((0,225000))
plt.xticks(range(0,70,4),values[range(0,70,4),1],rotation=45)
plt.title('2000-2017年各产业各季度生产总值散点图')
plt.legend(['第一产业','第二产业','第三产业'])
plt.savefig()
plt.show()
```
